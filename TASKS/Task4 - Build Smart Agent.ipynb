# Smart Agent (Mini AI Tool System)

This project demonstrates a simple rule-based AI agent that can:
- Fetch weather data
- Read files from local system
- Respond to user queries

It simulates how real AI agents use tools.

## Objective

- Build a basic AI agent system
- Add multiple tools (weather + file reader)
- Process user input commands
- Return meaningful outputs based on intent detection

```python
import os
import requests
```

## Weather Tool

This tool fetches real-time weather data using an API.

```python
def get_weather(city):
    api_key = "YOUR_API_KEY"  # replace with OpenWeather API key
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        temp = data["main"]["temp"]
        weather = data["weather"][0]["description"]
        return f"Weather in {city}: {temp}°C, {weather}"
    else:
        return "Weather data not found"
```

## File Reader Tool

This tool reads content from a local text file.

```python
def read_file(filename):
    if not os.path.exists(filename):
        return "File not found"
    
    with open(filename, "r", encoding="utf-8") as file:
        return file.read()
```

## Smart Agent Logic

The agent decides which tool to use based on user input.

```python
def smart_agent(user_input):
    user_input = user_input.lower()
    
    # Weather intent
    if "weather" in user_input:
        city = user_input.replace("check weather in", "").replace("weather in", "").strip()
        return get_weather(city)
    
    # File reading intent
    elif "read file" in user_input:
        filename = user_input.replace("read file", "").strip()
        return read_file(filename)
    
    else:
        return "Sorry, I don't understand the request."
```

## Example Input

Try:
- "Check weather in Bangalore"
- "Read file notes.txt"

```python
user_input = input("Enter your query: ")
result = smart_agent(user_input)

print("\nAgent Response:")
print(result)
```

## Conclusion

We built a simple Smart Agent that:
- Uses a weather API tool
- Reads files from system
- Detects intent from user input

This is a basic version of how AI agents like ChatGPT use tools internally.
