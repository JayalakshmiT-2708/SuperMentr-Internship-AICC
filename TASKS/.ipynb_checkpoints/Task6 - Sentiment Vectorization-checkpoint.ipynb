# Sentiment Vectorization Comparison

This notebook compares two text vectorization techniques:

1. Bag of Words (CountVectorizer)
2. TF-IDF (Term Frequency - Inverse Document Frequency)

We will:
- Convert sentences into vectors
- Display matrices
- Compare results visually using pandas

## Dataset

We use 4 simple sentiment sentences:
- Positive
- Negative
- Mixed sentiment

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
sentences = [
    "I love this movie",
    "This movie is terrible",
    "Amazing acting",
    "Worst film ever"
]
```

## 1. Bag of Words (CountVectorizer)

Counts how many times each word appears in a sentence.

```python
bow = CountVectorizer()

bow_matrix = bow.fit_transform(sentences)

bow_df = pd.DataFrame(
    bow_matrix.toarray(),
    columns=bow.get_feature_names_out()
)

bow_df.index = sentences

bow_df
```

## Bag of Words Output Meaning

- Each column = a word
- Each row = sentence
- Values = frequency of words in that sentence

## 2. TF-IDF Vectorization

TF-IDF gives importance to words based on:
- Frequency in sentence
- Rarity across dataset

```python
tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(sentences)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df.index = sentences

tfidf_df
```

## TF-IDF Output Meaning

- Common words like "this", "is" get lower weight
- Important words like "love", "terrible", "worst" get higher weight
- Better for sentiment understanding

## Bag of Words vs TF-IDF Comparison

We now compare both representations side by side.

```python
print("BAG OF WORDS MATRIX:\n")
print(bow_df)

print("\n\nTF-IDF MATRIX:\n")
print(tfidf_df)
```

## Visualization using Pandas

We display both matrices as tables for easy comparison.

```python
print("=== BAG OF WORDS ===")
display(bow_df)

print("\n=== TF-IDF ===")
display(tfidf_df)
```

## Conclusion

- Bag of Words counts word frequency
- TF-IDF highlights important words and reduces common word impact

👉 TF-IDF is generally better for sentiment analysis because it captures importance, not just frequency.
