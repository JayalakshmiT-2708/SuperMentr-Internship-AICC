# Car Price Prediction (Multiple Linear Regression)

This project predicts car price based on:
- Engine Size
- Mileage
- Age

We also analyze feature importance using model coefficients.

## Objective

- Create dataset
- Train Linear Regression model
- Predict car price for given input
- Print coefficients
- Interpret feature importance

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## Dataset

We create a small dataset with:
- Engine Size (cc)
- Mileage (km/l or similar)
- Age (years)
- Price (target variable)

```python
data = {
    "EngineSize": [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],
    "Mileage": [25, 22, 20, 18, 15, 14, 12, 10],
    "Age": [1, 2, 3, 4, 5, 6, 7, 8],
    "Price": [300000, 350000, 450000, 600000, 750000, 850000, 950000, 1100000]
}

df = pd.DataFrame(data)
df
```

## Train Model

```python
X = df[["EngineSize", "Mileage", "Age"]]
y = df["Price"]

model = LinearRegression()
model.fit(X, y)
```

## Predict Car Price

We predict price for:
- Engine Size = 1500
- Mileage = 20
- Age = 3

```python
pred = model.predict([[1500, 20, 3]])

print("Predicted Price:", pred[0])
```

## Model Coefficients

Coefficients show how each feature affects price.

```python
features = ["EngineSize", "Mileage", "Age"]

for feature, coef in zip(features, model.coef_):
    print(feature, ":", coef)

print("\nIntercept:", model.intercept_)
```

## Feature Importance Interpretation

- Positive coefficient → increases price
- Negative coefficient → decreases price

Expected:
- Engine Size → positive impact (higher engine = higher price)
- Mileage → negative impact (more mileage = lower price)
- Age → negative impact (older car = lower price)

## Conclusion

- Linear Regression models relationship between features and price
- Engine size is usually the most important positive factor
- Mileage and age reduce car value

This is a basic regression model used in real-world pricing systems.
