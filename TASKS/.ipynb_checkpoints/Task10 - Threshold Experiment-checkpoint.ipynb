# Threshold Experiment (Cancer Detection)

We explore how changing classification thresholds affects:

- Accuracy
- Precision
- Recall

We use the Breast Cancer dataset and a trained model.

## Objective

- Train a classifier (Logistic Regression)
- Get prediction probabilities
- Apply different thresholds (0.3, 0.5, 0.7)
- Compute accuracy, precision, recall
- Compare results in a table
- Decide best threshold for cancer detection

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

## Load Dataset

```python
data = load_breast_cancer()

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## Train Logistic Regression Model

We use Logistic Regression because it gives probability outputs.

```python
model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)

probs = model.predict_proba(X_test)[:, 1]  # probability of class 1
```

## Apply Different Thresholds

We test:
- 0.3 (low threshold → more positives)
- 0.5 (default)
- 0.7 (high threshold → stricter prediction)

```python
thresholds = [0.3, 0.5, 0.7]

results = []

for t in thresholds:
    preds = (probs >= t).astype(int)
    
    acc = accuracy_score(y_test, preds)
    prec = precision_score(y_test, preds)
    rec = recall_score(y_test, preds)
    
    results.append([t, acc, prec, rec])
```

```python
df_results = pd.DataFrame(results, columns=["Threshold", "Accuracy", "Precision", "Recall"])
df_results
```

## Observations

- Lower threshold → more positive predictions
- Higher recall but lower precision
- Higher threshold → fewer positives
- Higher precision but lower recall

## Which threshold is best for cancer detection?

👉 Best threshold: 0.3 or slightly below 0.5

### Because

In cancer detection:
- Missing a positive case (False Negative) is very dangerous
- We want HIGH RECALL (catch all possible cancer cases)

So:
✔ Lower threshold increases recall  
✔ Better to flag more cases for medical review  
✔ False positives are acceptable, but false negatives are not
