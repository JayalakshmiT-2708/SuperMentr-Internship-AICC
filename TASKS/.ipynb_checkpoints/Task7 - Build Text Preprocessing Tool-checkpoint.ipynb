# Text Preprocessing Tool (NLP Pipeline)

This program demonstrates basic NLP preprocessing steps:

- Lowercasing
- Removing punctuation
- Tokenization
- Stopword removal
- Stemming

These steps are essential before training any NLP model.

## Objective

To clean raw text and convert it into meaningful tokens for NLP tasks.

Input:
"I am learning NLP and it is very exciting!!!"

Output:
['learn', 'nlp', 'excit']

```python
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = "I am learning NLP and it is very exciting!!!"
print("Original Text:", text)
```

## Step 1: Convert to Lowercase

We convert all text into lowercase to ensure uniformity.

```python
text = text.lower()
print(text)
```

## Step 2: Remove Punctuation

We remove punctuation marks like !, ., ?, etc.

```python
text = text.translate(str.maketrans("", "", string.punctuation))
print(text)
```

## Step 3: Tokenization

We split text into individual words (tokens).

```python
tokens = word_tokenize(text)
print(tokens)
```

## Step 4: Remove Stopwords

Stopwords are common words like "is", "am", "it", which do not carry meaning.

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens)
```

## Step 5: Stemming

Stemming reduces words to their root form.
Example: "exciting" → "excit"

```python
stemmer = PorterStemmer()

stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print(stemmed_tokens)
```

## Final Output

The cleaned and processed tokens are:

```python
print(stemmed_tokens)
```

## Conclusion

We successfully built an NLP preprocessing pipeline that:
- Cleans raw text
- Removes noise
- Converts words into meaningful root forms

This is a fundamental step in all NLP and ML text-based systems.
