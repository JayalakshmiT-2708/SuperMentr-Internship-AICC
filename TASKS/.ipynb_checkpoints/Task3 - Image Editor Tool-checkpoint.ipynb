# Image Editor Tool (Mini OpenCV App)

This project demonstrates a simple image processing tool using OpenCV.

It can:
- Load an image
- Convert to grayscale
- Apply blur
- Detect edges

## Objective

To understand basic image processing operations using OpenCV:

- Image loading
- Grayscale conversion
- Blurring (smoothing)
- Edge detection

This helps in building the foundation for computer vision applications.

```python
import cv2
import matplotlib.pyplot as plt
```

## Load Image

We load an image and convert it from BGR (OpenCV default) to RGB for display.

```python
# Load image (replace with your image path)
image = cv2.imread("image.jpg")

# Convert BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

## Convert to Grayscale

Grayscale removes color information and keeps only intensity values.

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

## Apply Blur

Blurring helps reduce noise using averaging filter.

```python
blur = cv2.GaussianBlur(image, (7, 7), 0)

blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

plt.imshow(blur_rgb)
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

## Edge Detection

We use Canny Edge Detection to highlight edges in the image.

```python
edges = cv2.Canny(image, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

## Conclusion

We successfully built a simple image editor using OpenCV.

### Operations performed:
- Image loading
- Grayscale conversion
- Blurring
- Edge detection

This is the foundation of computer vision systems used in AI applications.
