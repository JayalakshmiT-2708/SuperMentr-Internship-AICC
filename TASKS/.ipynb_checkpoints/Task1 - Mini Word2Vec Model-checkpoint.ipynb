# Mini Word2Vec Model

This notebook demonstrates training a simple Word2Vec model on a custom dataset and exploring word embeddings.

## Objective

- Create a small dataset of sentences (college life topic)
- Train a Word2Vec model
- Display vector representation of a word
- Find most similar words

This helps in understanding how word embeddings are learned from context.

```python
%pip install gensim
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
```

## Dataset

We create a small dataset of 15 sentences related to college life.

```python
sentences = [
    "college life is full of fun and learning",
    "students attend lectures and complete assignments",
    "friends study together before exams",
    "campus has a library and cafeteria",
    "group projects help in teamwork",
    "exams create stress for students",
    "teachers guide students in academics",
    "students participate in cultural events",
    "college festivals are exciting",
    "hostel life teaches independence",
    "students prepare presentations for class",
    "internships help students gain experience",
    "late night study is common during exams",
    "students enjoy breaks between classes",
    "college life builds memories and friendships"
]
```

## Preprocessing

We tokenize each sentence into words and convert them to lowercase.

```python
data = []

for sentence in sentences:
    tokens = word_tokenize(sentence.lower())
    data.append(tokens)

print(data[:3])
```

## Training Word2Vec Model

We train the Word2Vec model using:
- vector_size = 50
- window = 3
- min_count = 1

```python
model = Word2Vec(
    sentences=data,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

## Word Vector Representation

We extract the vector for a specific word.

```python
word = "students"
vector = model.wv[word]

print(f"Vector for '{word}':\n", vector)
```

## Most Similar Words

We find words that are similar based on context.

```python
similar_words = model.wv.most_similar("students", topn=5)

print("Most similar words to 'students':")
for word, score in similar_words:
    print(word, ":", score)
```

## Conclusion

- Word2Vec converts words into numerical vectors.
- Words used in similar contexts have similar meanings.
- The model learns relationships automatically from data.

This demonstrates how machines understand language contextually.
