# Student Performance Analysis

This project analyzes student marks from a CSV file.

We will:
- Load CSV data
- Handle missing values
- Add Total and Average columns
- Find Top 3 students
- Compute department-wise averages
- Filter high-performing students

## Objective

To perform data preprocessing and analysis using pandas:

- Data cleaning
- Feature engineering (Total, Average)
- Group analysis
- Filtering based on conditions

```python
import pandas as pd
```

## Load Dataset

We read the CSV file containing student marks.

```python
import pandas as pd
import numpy as np

# Creating data that fits your specific analysis
data = {
    'student_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'name': ['Arjun', 'Sanya', 'Vikram', 'Priya', 'Rohan', 'Anjali', 'Karan', 'Meera'],
    'dept': ['CS', 'CS', 'ECE', 'MECH', 'ECE', 'CS', 'MECH', 'ECE'],
    'maths': [85, 92, np.nan, 45, 88, 76, 90, 82],      # One missing value
    'science': [78, np.nan, 85, 30, 92, 80, 88, 77],    # One missing value
    'english': [90, 85, 70, 50, np.nan, 82, 95, 79]     # One missing value
}

df_new = pd.DataFrame(data)

# Save to CSV
df_new.to_csv('students.csv', index=False)

print("New 'students.csv' created with 'dept' and subject columns!")
```

```python
df = pd.read_csv("students.csv")

df
```

## Handle Missing Values

We check for missing values and fill them with 0 (or mean if required).

```python
print(df.isnull().sum())

df.fillna(0, inplace=True)
```

## Add Total and Average Columns

```python
df["total"] = df["maths"] + df["science"] + df["english"]

df["average"] = df["total"] / 3

df
```

## Top 3 Students Based on Total Marks

```python
top3 = df.sort_values(by="total", ascending=False).head(3)

top3
```

## Department-wise Average Marks

```python
dept_avg = df.groupby("dept")[["maths", "science", "english", "total", "average"]].mean()

dept_avg
```

## Students Scoring Above 75 in All Subjects

```python
high_performers = df[
    (df["maths"] > 75) &
    (df["science"] > 75) &
    (df["english"] > 75)
]

high_performers
```
