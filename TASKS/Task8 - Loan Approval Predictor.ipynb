# Loan Approval Predictor

This project predicts whether a loan will be approved based on applicant details.

We use:
- Decision Tree
- Random Forest

We compare both models and analyze feature importance.

## Objective

- Load and preprocess dataset
- Train Decision Tree and Random Forest models
- Compare accuracy
- Identify important features
- Save trained model using pickle

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
```

## Dataset

We assume a dataset with the following columns:

- Income
- Credit Score
- Age
- Loan Amount
- Employment Years
- Loan Approved (Target: 0 or 1)

```python
data = {
    "Income": [50000, 60000, 25000, 80000, 120000, 30000, 70000, 90000, 40000, 100000],
    "CreditScore": [650, 700, 550, 800, 850, 500, 720, 780, 600, 820],
    "Age": [25, 30, 22, 45, 50, 23, 35, 40, 28, 48],
    "LoanAmount": [200000, 250000, 100000, 300000, 500000, 120000, 220000, 350000, 150000, 400000],
    "EmploymentYears": [2, 5, 1, 10, 15, 1, 6, 8, 3, 12],
    "LoanApproved": [0, 1, 0, 1, 1, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

## Train-Test Split

We split the dataset into training and testing sets.

```python
X = df.drop("LoanApproved", axis=1)
y = df["LoanApproved"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## Decision Tree Model

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

dt_pred = dt_model.predict(X_test)

dt_acc = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_acc)
```

## Random Forest Model

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_acc)
```

## Model Comparison

We compare accuracy of both models.

```python
print("Decision Tree Accuracy:", dt_acc)
print("Random Forest Accuracy:", rf_acc)
```

## Feature Importance (Random Forest)

We check which features influence loan approval most.

```python
importance = rf_model.feature_importances_

features = X.columns

for feature, score in zip(features, importance):
    print(feature, ":", score)
```

## Save Model using Pickle

We save the trained Random Forest model for future use.

```python
with open("loan_model.pkl", "wb") as file:
    pickle.dump(rf_model, file)

print("Model saved successfully!")
```

## Load Saved Model

```python
with open("loan_model.pkl", "rb") as file:
    loaded_model = pickle.load(file)

sample = [[70000, 720, 30, 200000, 5]]

prediction = loaded_model.predict(sample)

print("Prediction:", prediction[0])
```

## Conclusion

- Random Forest generally performs better than Decision Tree
- Credit Score and Income are strong predictors
- Model is saved and ready for deployment

This project demonstrates a full ML pipeline from training to deployment.
