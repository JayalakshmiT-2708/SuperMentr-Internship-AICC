# Smarter Text Generator (Mini LLM)

This notebook builds a simple text generator using Word2Vec.
It takes user input and generates a 10-word sentence based on learned word relationships.

## Objective

- Train a Word2Vec model
- Accept user input word
- Predict next words using similarity
- Generate a 10-word sentence

This simulates a very basic version of how language models generate text.

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
import random

nltk.download('punkt')
```

## Dataset

We use a small dataset based on college life.

```python
sentences = [
    "college life is full of fun and learning",
    "students attend lectures and complete assignments",
    "friends study together before exams",
    "campus has a library and cafeteria",
    "group projects help in teamwork",
    "exams create stress for students",
    "teachers guide students in academics",
    "students participate in cultural events",
    "college festivals are exciting",
    "hostel life teaches independence",
    "students prepare presentations for class",
    "internships help students gain experience",
    "late night study is common during exams",
    "students enjoy breaks between classes",
    "college life builds memories and friendships"
]
```

## Preprocessing

Convert sentences into tokenized form.

```python
data = []

for sentence in sentences:
    tokens = word_tokenize(sentence.lower())
    data.append(tokens)
```

## Train Word2Vec Model

```python
model = Word2Vec(
    sentences=data,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

## Text Generation Logic

- Take a starting word from user
- Find most similar words
- Randomly pick next word from top similar words
- Repeat until 10 words are generated

```python
def generate_sentence(start_word, model, length=10):
    sentence = [start_word]

    for _ in range(length - 1):
        try:
            similar_words = model.wv.most_similar(sentence[-1], topn=5)
            next_word = random.choice(similar_words)[0]
            sentence.append(next_word)
        except KeyError:
            break

    return " ".join(sentence)
```

## User Input

Enter a starting word to generate a sentence.

```python
start_word = input("Enter a starting word: ").lower()

generated_text = generate_sentence(start_word, model, length=10)

print("\nGenerated Sentence:")
print(generated_text)
```

## Conclusion

- This mini model generates text using word similarity
- It does not understand grammar fully
- But it captures contextual relationships between words

This is a simplified version of how larger language models work.
