## Introduction

Semantic similarity refers to how closely related the meanings of two words are.

In Natural Language Processing (NLP), understanding semantic similarity helps in tasks like search engines, chatbots, and recommendation systems.

## Word Pairs with Semantic Similarity

### 1. Car – Automobile
These words are highly similar as both refer to the same type of vehicle. They are synonyms.

### 2. Big – Large
Both words describe size and have almost identical meanings. They are commonly interchangeable.

### 3. Happy – Joyful
These words express positive emotions and are closely related in meaning.

### 4. Teacher – Professor
Both refer to people who teach, though "professor" is more specific (usually in a college or university).

### 5. Dog – Animal
These words are related, but not identical. A dog is a type of animal, so the similarity is hierarchical (category relationship).

## Types of Semantic Relationships

1. Synonyms:
   - Words with similar meanings (e.g., Big – Large)

2. Antonyms:
   - Words with opposite meanings (e.g., Hot – Cold)

3. Hypernym:
   - General category (e.g., Animal)

4. Hyponym:
   - Specific instance (e.g., Dog)

5. Related Words:
   - Words connected by context (e.g., Teacher – Student)

## Conclusion

Semantic similarity helps machines understand relationships between words. It plays an important role in improving the accuracy of NLP applications.
