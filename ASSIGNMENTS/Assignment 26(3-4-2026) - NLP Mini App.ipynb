## Introduction

Keyword extraction is an NLP technique used to identify important words or phrases from text.

In this mini application, we build a keyword extractor using TF-IDF to identify the most relevant words from a given input.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def extract_keywords(text):
    vectorizer = TfidfVectorizer(stop_words='english')
    
    X = vectorizer.fit_transform([text])
    
    feature_names = vectorizer.get_feature_names_out()
    scores = X.toarray()[0]
    
    word_scores = list(zip(feature_names, scores))
    sorted_words = sorted(word_scores, key=lambda x: x[1], reverse=True)
    
    return sorted_words[:5]
```

```python
text = "Machine learning is a powerful tool for data analysis and prediction."

keywords = extract_keywords(text)

print("Top Keywords:")
for word, score in keywords:
    print(word)
```

```python
texts = [
    "Artificial intelligence is transforming the world",
    "Data science includes statistics and machine learning",
    "Natural language processing helps computers understand text"
]

for t in texts:
    print("\nText:", t)
    print("Keywords:", [word for word, _ in extract_keywords(t)])
```

## Explanation

- TF-IDF is used to calculate the importance of words
- Common words (like "is", "the") are removed using stopwords
- Words with higher scores are considered keywords

## Conclusion

This mini NLP application extracts important keywords from text using TF-IDF.

It can be used in search engines, document analysis, and summarization systems.
