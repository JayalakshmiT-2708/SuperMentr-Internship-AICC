```python
# Student Data Manager
students = {
    "Aarav": 85,
    "Bhavya": 92,
    "Charan": 78,
    "Divya": 88,
    "Esha": 95
}

# Finding the topper
topper = max(students, key=students.get)
topper_marks = students[topper]

# Calculating class average
total_marks = sum(students.values())
average_marks = total_marks / len(students)

# Assigning grades
def assign_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "D"

# Displaying results
print("Student Marks:", students)
print("Topper:", topper, "with marks", topper_marks)
print("Class Average:", average_marks)

print("\nGrades:")
for student, marks in students.items():
    print(student, ":", assign_grade(marks))
```
