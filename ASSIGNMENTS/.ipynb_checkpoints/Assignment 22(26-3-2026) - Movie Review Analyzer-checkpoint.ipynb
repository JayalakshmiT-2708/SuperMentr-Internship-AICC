## Introduction

Sentiment Analysis is a Natural Language Processing (NLP) technique used to determine whether a piece of text expresses a positive, negative, or neutral sentiment.

In this assignment, we build a simple sentiment analyzer and test it on movie reviews.

```python
%pip install textblob
```

```python
import nltk
nltk.download('punkt')
from textblob import TextBlob
```

```python
def analyze_sentiment(text):
    analysis = TextBlob(text)
    
    polarity = analysis.sentiment.polarity
    
    if polarity > 0:
        return "Positive 😊"
    elif polarity < 0:
        return "Negative 😡"
    else:
        return "Neutral 😐"
```

```python
reviews = [
    "This movie was absolutely amazing, I loved it!",
    "The film was boring and too long.",
    "It was an average movie, nothing special.",
    "Fantastic acting and great story!",
    "Worst movie ever, complete waste of time."
]
```

```python
for review in reviews:
    result = analyze_sentiment(review)
    
    print("Review:", review)
    print("Sentiment:", result)
    print("-" * 50)
```

## Explanation

- The model uses TextBlob to calculate polarity:
  - Positive (> 0)
  - Negative (< 0)
  - Neutral (= 0)

- Based on polarity, the sentiment is classified.

Example:
- "amazing" → Positive
- "boring" → Negative

## Conclusion

Sentiment analysis helps in understanding user opinions from text data.

This simple model can classify movie reviews into positive, negative, or neutral categories. It can be further improved using advanced machine learning models.
