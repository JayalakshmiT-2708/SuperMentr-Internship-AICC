## Introduction

A Decision Tree is a supervised machine learning algorithm used for classification and decision-making.

It works like a flowchart where each decision leads to another question or a final outcome.

In this assignment, we design a decision tree to decide whether to play outside based on weather conditions.

## Features Considered

We use the following features to make the decision:

1. Weather (Sunny, Rainy, Cloudy)
2. Temperature (Hot, Mild, Cool)
3. Humidity (High, Normal)
4. Wind (Strong, Weak)

## Explanation of the Decision Tree

- If the weather is Sunny:
  - Check humidity:
    - If High → Do not play
    - If Normal → Play

- If the weather is Rainy:
  - Check wind:
    - If Strong → Do not play
    - If Weak → Play

- If the weather is Cloudy:
  - Always Play
