## Introduction

House price prediction is a common machine learning problem where we predict the price of a house based on features like area, number of rooms, etc.

In this assignment, we use Linear Regression to train a model and predict house prices.

## What is Linear Regression?

Linear Regression is a supervised machine learning algorithm used to predict a continuous value.

It finds a relationship between input (X) and output (Y) using a straight line.

### Equation:
Y = mX + c

Where:
- Y = Predicted value
- X = Input feature
- m = Slope
- c = Intercept

y = mx + c

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Example dataset (Area vs Price)

# Area in sq.ft
X = np.array([[500], [1000], [1500], [2000], [2500]])

# Price in lakhs
y = np.array([50, 100, 150, 200, 250])

print("Area:\n", X)
print("Prices:\n", y)
```

```python
# Create model
model = LinearRegression()

# Train model
model.fit(X, y)
```

```python
# Slope (m)
print("Slope (m):", model.coef_)

# Intercept (c)
print("Intercept (c):", model.intercept_)
```

```python
# Predict price for new house (e.g., 1800 sq.ft)
new_area = np.array([[1800]])

predicted_price = model.predict(new_area)

print("Predicted Price:", predicted_price)
```

```python
# Test with multiple inputs
test_areas = np.array([[1200], [2200], [3000]])

predictions = model.predict(test_areas)

for area, price in zip(test_areas, predictions):
    print(f"Area: {area[0]} sq.ft -> Price: {price}")
```

## Conclusion

Linear Regression is a simple and effective algorithm for predicting continuous values like house prices.

In this assignment, we trained a model using area as input and predicted house prices for new data.

This approach is widely used in real estate and business applications.
