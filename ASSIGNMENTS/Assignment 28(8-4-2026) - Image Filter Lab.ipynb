## Introduction

Image processing is an important part of computer vision. It involves modifying images to extract useful information.

In this assignment, we apply different filters such as grayscale, blur, and edge detection using OpenCV.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('image.jpg')

# Convert to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)

plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
```

```python
edges = cv2.Canny(blur, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis('off')
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,4,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(blur, cmap='gray')
plt.title("Blur")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(edges, cmap='gray')
plt.title("Edges")
plt.axis('off')

plt.show()
```

## Explanation

### 1. Grayscale
- Converts image to black and white
- Reduces complexity (3 channels → 1 channel)

### 2. Blur
- Smoothens the image
- Removes noise and small details

### 3. Edge Detection
- Detects boundaries of objects
- Highlights important structures in image

## Conclusion

Image filters help in preprocessing and feature extraction in computer vision.

Grayscale simplifies the image, blur removes noise, and edge detection highlights important features.
