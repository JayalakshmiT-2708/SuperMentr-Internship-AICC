## Introduction

In computer vision, images are represented as numerical data. Each image is made up of pixels, and each pixel contains intensity values.

In this assignment, we load an image and explore its shape, pixel values, and color channels.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image (make sure image is in same folder)
img = cv2.imread('image.jpg')

# Convert BGR to RGB (important for display)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Loaded Image")
plt.axis('off')
```

```python
print("Image Shape:", img.shape)
```

```python
print("Pixel Value at (0,0):", img[0, 0])
```

```python
# Split channels
b, g, r = cv2.split(img)

print("Blue channel shape:", b.shape)
print("Green channel shape:", g.shape)
print("Red channel shape:", r.shape)
```

## Explanation

### 1. Image Shape
The shape of the image is represented as:
(height, width, channels)

- Height → number of rows
- Width → number of columns
- Channels → color components (usually 3 for RGB)

Example: (300, 400, 3)

---

### 2. Pixel Values
Each pixel contains intensity values:
- Range: 0 to 255
- Example: [255, 0, 0] represents a color (Blue/Green/Red depending on format)

---

### 3. Channels
A color image has 3 channels:
- Red
- Green
- Blue

Each channel represents intensity of that color.

## Conclusion

Images are numerical representations of pixel values. Understanding shape, pixel values, and channels is essential for image processing and computer vision tasks.
