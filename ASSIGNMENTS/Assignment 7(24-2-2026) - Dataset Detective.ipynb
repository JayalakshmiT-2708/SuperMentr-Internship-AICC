```python
import pandas as pd
import os

# Check files in current folder (for debugging)
print("Files in current directory:")
print(os.listdir())

# Load dataset
data = pd.read_csv("dataset.csv")

# Display top 5 rows
print("\nTop rows of the dataset:")
print(data.head())

# Get numeric columns
numeric_columns = data.select_dtypes(include='number')

# Find highest value and its column
highest_values = numeric_columns.max()
highest_value = highest_values.max()
highest_column = highest_values.idxmax()

print("\nHighest value in dataset:", highest_value)
print("Column with highest value:", highest_column)

# Count missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```
