## Introduction

TF-IDF (Term Frequency - Inverse Document Frequency) is a technique used in Natural Language Processing to identify important words in a document.

It helps highlight words that are frequent in a document but rare across multiple documents.

## What is TF-IDF?

TF-IDF is calculated as:

TF-IDF = TF × IDF

- TF (Term Frequency): How often a word appears in a document
- IDF (Inverse Document Frequency): How rare the word is across documents

Words with high TF-IDF scores are considered important keywords.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
documents = [
    "Machine learning is very useful in real world applications",
    "Artificial intelligence and machine learning are closely related",
    "Deep learning is a subset of machine learning",
    "Natural language processing is a part of artificial intelligence",
    "Data science uses machine learning and statistics"
]
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

df = pd.DataFrame(X.toarray(), columns=feature_names)

df
```

```python
for i, doc in enumerate(df.values):
    print(f"\nDocument {i+1} Top Words:")
    
    word_scores = list(zip(feature_names, doc))
    sorted_words = sorted(word_scores, key=lambda x: x[1], reverse=True)
    
    for word, score in sorted_words[:5]:
        print(word, ":", score)
```

## Explanation of Results

- Words like "machine", "learning", and "data" appear frequently, but their importance depends on how unique they are in each document.
- Words that are common across all documents get lower TF-IDF scores.
- Words that are unique to a specific document get higher scores.

Example:
- "statistics" may have a high score in one document because it appears less frequently in others.

## Conclusion

TF-IDF is a powerful technique to identify important words in text data. It helps in keyword extraction, search engines, and text analysis.

This method improves the understanding of document importance in NLP tasks.
