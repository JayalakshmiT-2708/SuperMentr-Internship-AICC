## Introduction

Prompt engineering is the process of designing effective inputs to get better outputs from AI models.

A well-structured prompt produces more accurate, detailed, and useful results compared to a vague prompt.

## 1. Resume Prompt

### Weak Prompt:
"Make a resume for me"

### Strong Prompt:
"Create a professional resume for a Computer Science student with skills in Python, Machine Learning, and basic web development. Include sections for education, projects, skills, and certifications."

### Comparison:
- Weak prompt is vague and lacks details
- Strong prompt gives context, role, and structure

## 2. Business Idea Prompt

### Weak Prompt:
"Give me a business idea"

### Strong Prompt:
"Suggest a unique business idea for a college student with low investment, focusing on AI or technology, and explain how it can generate profit."

### Comparison:
- Weak prompt gives no direction
- Strong prompt specifies budget, domain, and expectation

## 3. Study Plan Prompt

### Weak Prompt:
"Give me a study plan"

### Strong Prompt:
"Create a 2-month study plan for a beginner to learn Machine Learning, including daily topics, practice exercises, and project ideas."

### Comparison:
- Weak prompt is unclear
- Strong prompt defines duration, level, and goals

## Key Observations

- Strong prompts provide clear instructions
- They include context, constraints, and goals
- They result in more useful and accurate outputs
- Weak prompts are vague and produce generic responses

## Conclusion

Prompt engineering plays a crucial role in interacting with AI systems. By writing clear and detailed prompts, users can significantly improve the quality of AI-generated responses.
