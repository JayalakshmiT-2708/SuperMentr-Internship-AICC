## Introduction

Text preprocessing is an important step in Natural Language Processing (NLP). It helps clean and standardize text data for better analysis.

In this assignment, we build a text cleaner that:
- Converts text to lowercase
- Removes punctuation
- Removes stopwords

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def clean_text(text):
    # 1. Convert to lowercase
    text = text.lower()
    
    # 2. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 3. Tokenize text
    words = word_tokenize(text)
    
    # 4. Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    
    # 5. Join words back
    cleaned_text = ' '.join(filtered_words)
    
    return cleaned_text
```

```python
sample_text = "OMG!!! This is soooo amazing 😍🔥, I can't believe it!!!"

cleaned = clean_text(sample_text)

print("Original Text:")
print(sample_text)

print("\nCleaned Text:")
print(cleaned)
```

```python
texts = [
    "Hello!!! How are you???",
    "This is a simple TEST sentence.",
    "I am veryyyy happppy!!! 😄",
    "Please send me the notes ASAP!!!"
]

for t in texts:
    print("Original:", t)
    print("Cleaned:", clean_text(t))
    print("-" * 40)
```

## Conclusion

Text cleaning is a crucial preprocessing step in NLP. By removing punctuation, converting text to lowercase, and eliminating stopwords, we can improve the quality of text data for machine learning models.
