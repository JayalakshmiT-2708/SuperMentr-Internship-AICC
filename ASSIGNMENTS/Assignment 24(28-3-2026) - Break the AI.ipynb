## Introduction

Large Language Models (LLMs) are powerful AI systems used for generating human-like text. However, they can sometimes be confused by ambiguous, tricky, or misleading prompts.

In this assignment, we test an AI model using challenging prompts and observe its behavior.

## Tricky Prompts and Results

### 1. Ambiguous Question
**Prompt:**  
"Can a man marry his widow's sister?"

**AI Response:**  
No, because if he has a widow, he is already dead.

**Observation:**  
The question is logically tricky. The AI correctly identifies the contradiction.

---

### 2. Word Confusion
**Prompt:**  
"What is heavier: 1 kg of cotton or 1 kg of iron?"

**AI Response:**  
Both are equal.

**Observation:**  
This tests basic reasoning. AI answers correctly, but many humans get confused.

---

### 3. Contradictory Instruction
**Prompt:**  
"Write a sentence that is both completely true and completely false."

**AI Response:**  
This sentence is false.

**Observation:**  
This creates a paradox (liar paradox), confusing logical consistency.

---

### 4. Misleading Context
**Prompt:**  
"I put a glass on a table and then turned the table upside down. Where is the glass?"

**AI Response:**  
The glass would likely fall off.

**Observation:**  
AI assumes real-world physics, but question lacks constraints.

---

### 5. Trick Grammar
**Prompt:**  
"Buffalo buffalo Buffalo buffalo buffalo buffalo Buffalo buffalo."

**AI Response:**  
Struggles to interpret correctly.

**Observation:**  
Grammatically valid but extremely confusing sentence.

## Analysis

From the above examples, we observe:

- AI can handle logical traps but may struggle with ambiguity.
- Paradoxical statements create confusion.
- Complex grammar can reduce understanding.
- AI relies on patterns and may assume missing information.

## Conclusion

Although LLMs are powerful, they are not perfect. They can be confused by tricky prompts, ambiguous questions, and paradoxes.

Understanding these limitations helps improve AI systems and their reliability.
